```python
import os
from datasets import load_dataset

voxpopuli = load_dataset("esb/datasets", "voxpopuli", trust_remote_code=True, token=os.environ['HS_TOKEN'])
cv_13 = load_dataset("mozilla-foundation/common_voice_13_0", "en", trust_remote_code=True, token=os.environ['HS_TOKEN'])
```

```python
voxpopuli['train'][0]
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

tokenizer = AutoTokenizer.from_pretrained("google/gemma-7b-it", token=os.environ['HS_TOKEN'])
model = AutoModelForCausalLM.from_pretrained(
    "google/gemma-7b-it",
    torch_dtype=torch.bfloat16,
    token=os.environ['HS_TOKEN']
)
```

```python
voxpopuli = voxpopuli.remove_columns(['id', 'audio', 'dataset'])
cv_13 = cv_13.remove_columns(['client_id', 'path', 'up_votes', 'down_votes', 'age', 'gender', 'accent', 'locale', 'segment', 'variant'])
cv_13 = cv_13.rename_column('sentence', 'text')
```
